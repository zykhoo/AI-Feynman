```python
def replace_equations(input_string):
  input_string = input_string.replace("exp", "np.exp")
  input_string = input_string.replace("sqrt", "np.sqrt")
  input_string = input_string.replace("pi", "np.pi")
  input_string = input_string.replace("log", "np.log")
  input_string = input_string.replace("sin", "np.sin")
  input_string = input_string.replace("anp.sin", "np.arcsin")
  input_string = input_string.replace("cos", "np.cos")
  input_string = input_string.replace("anp.cos", "np.arccos")
  input_string = input_string.replace("tan", "np.tan")
  input_string = input_string.replace("anp.tan", "np.arctan")
  return input_string
```

# Cartesian

```python
cat /content/drive/MyDrive/mars/iiWAS2023/centricity/Results/areocentricity-cartesian.txt
```

```python
import pandas as pd
df = pd.DataFrame(columns=["equation", "fit", "parsimony", "colour"])
```

```python
with open("/content/drive/MyDrive/mars/iiWAS2023/centricity/Results/areocentricity-cartesian2.txt", "r") as file:
  data = file.readlines()
  count=0
  for j in data:
    count+=1
    equations = j.split(' ',5)
    equation = equations[-1]
    fit = equations[0]
    parsimony = equations[-3]
    equation = replace_equations(equation.replace("\n", ""))
    if ("x0" in equation)|("x1" in equation)|("x2" in equation)|("x3" in equation)|("x4" in equation)|("x5" in equation):
      equation = equation.replace("x0","MS_x")
      equation = equation.replace("x1","MS_y")
      equation = equation.replace("x2","MS_z")
      equation = equation.replace("x3","ME_x")
      equation = equation.replace("x4","ME_y")
      equation = equation.replace("x5","ME_z")
      print("Equation %s: %s, %s, %s" %(count, equation, fit, parsimony))
      entry = pd.DataFrame.from_dict({"equation": [equation], "fit": [float(fit)], "parsimony": [float(parsimony)], "colour": ["red"]})
      df = pd.concat([df, entry], ignore_index=True)
```

```python
with open("/content/drive/MyDrive/mars/iiWAS2023/centricity/Results/heliocentricity-cartesian2.txt", "r") as file:
  data = file.readlines()
  count=0
  for j in data:
    count+=1
    equations = j.split(' ',5)
    equation = equations[-1]
    fit = equations[0]
    parsimony = equations[-3]
    equation = replace_equations(equation.replace("\n", ""))
    if ("x0" in equation)|("x1" in equation)|("x2" in equation)|("x3" in equation)|("x4" in equation)|("x5" in equation):
      equation = equation.replace("x0","SM_x")
      equation = equation.replace("x1","SM_y")
      equation = equation.replace("x2","SM_z")
      equation = equation.replace("x3","SE_x")
      equation = equation.replace("x4","SE_y")
      equation = equation.replace("x5","SE_z")
      print("Equation %s: %s, %s, %s" %(count, equation, fit, parsimony))
      entry = pd.DataFrame.from_dict({"equation": [equation], "fit": [float(fit)], "parsimony": [float(parsimony)], "colour": ["green"]})
      df = pd.concat([df, entry], ignore_index=True)
```
