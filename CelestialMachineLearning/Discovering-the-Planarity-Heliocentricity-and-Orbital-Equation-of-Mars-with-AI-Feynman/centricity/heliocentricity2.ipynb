```python
from astroquery.jplhorizons import Horizons
import numpy as np
import plotly.graph_objects as go

# Get the ephemeris for the Sun and Mar
mars = Horizons(id='499', location='675', epochs={'start': '1601-01-01', 'stop': '1602-12-31', 'step': '1d'}) #1601-12-31
sun = Horizons(id='Sun', location='675', epochs={'start': '1601-01-01', 'stop': '1602-12-31', 'step': '1d'})

mars_epth = mars.ephemerides()
sun_epth = sun.ephemerides()
```

```python
from astropy.coordinates import SkyCoord
from astropy import units as u

mars_ang_width = mars_epth['ang_width']
sun_ang_width = sun_epth['ang_width']

mars_ra = mars_epth['RA']
mars_dec = mars_epth['DEC']

sun_ra = sun_epth['RA']
sun_dec = sun_epth['DEC']
```

# Cartesian

```python
m = SkyCoord(np.array(mars_ra)*u.degree, np.array(mars_dec)*u.degree, 0.00486759341* 0.00465047 / np.tan(np.radians(mars_ang_width / 3600) / 2))
m.represent_as('cartesian')
s = SkyCoord(np.array(sun_ra)*u.degree, np.array(sun_dec)*u.degree, 0.00465047 / np.tan(np.radians(sun_ang_width / 3600) / 2))
s.represent_as('cartesian')
```

```python
s_m = SkyCoord(-s.represent_as('cartesian').x + m.represent_as('cartesian').x,
               -s.represent_as('cartesian').y + m.represent_as('cartesian').y,
               -s.represent_as('cartesian').z + m.represent_as('cartesian').z, representation_type='cartesian') #view of mars from sun
s_e = SkyCoord(-s.represent_as('cartesian').x,
               -s.represent_as('cartesian').y,
               -s.represent_as('cartesian').z, representation_type='cartesian') #view of earth from sun
```

```python
m = SkyCoord(np.array(mars_ra)*u.degree, np.array(mars_dec)*u.degree, 0.00486759341* 0.00465047 / np.tan(np.radians(mars_ang_width / 3600) / 2))
m.represent_as('cartesian')
s = SkyCoord(np.array(sun_ra)*u.degree, np.array(sun_dec)*u.degree, 0.00465047 / np.tan(np.radians(sun_ang_width / 3600) / 2))
s.represent_as('cartesian')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(x=s_m.x, y=s_m.y, z=s_m.z, mode='markers',  marker_color='red')])
fig.add_trace(go.Scatter3d(x=s_e.x, y=s_e.y, z=s_e.z, mode='markers', marker_color='blue'))
fig.show()

```

```python
import os
import pandas as pd

# os.chdir("./AI-Feynman/") #&& ./compile.sh

cart_dist = np.sqrt((s_m.x)**2 + (s_m.y)**2 + (s_m.z)**2)


data_cartesian = pd.DataFrame()
data_cartesian['mars_x'] = s_m.x
data_cartesian['mars_y'] = s_m.y
data_cartesian['mars_z'] = s_m.z
data_cartesian['earth_x'] = s_e.x
data_cartesian['earth_y'] = s_e.y
data_cartesian['earth_z'] = s_e.z
data_cartesian['dist'] = cart_dist

data_cartesian = data_cartesian.sample(frac=1).reset_index(drop=True)

np.savetxt('./example_data/cartesian.txt', data_cartesian[[ "mars_x","mars_y","mars_z","earth_x","earth_y","earth_z","dist"]].sample(frac=1).values)
np.loadtxt('./example_data/cartesian.txt')
```

# Polar


```python
m = SkyCoord(np.array(mars_ra)*u.degree, np.array(mars_dec)*u.degree, 0.00486759341* 0.00465047 / np.tan(np.radians(mars_ang_width / 3600) / 2))
m.represent_as('cartesian')
s = SkyCoord(np.array(sun_ra)*u.degree, np.array(sun_dec)*u.degree, 0.00465047 / np.tan(np.radians(sun_ang_width / 3600) / 2))
s.represent_as('cartesian')
```

```python
from astropy.coordinates.representation import SphericalRepresentation

s_e = (-s.represent_as('cartesian')).represent_as(SphericalRepresentation)
s_m = (-s.represent_as('cartesian') + m.represent_as('cartesian')).represent_as(SphericalRepresentation)
```

```python
s_e[:10]
```

```python
import os
import pandas as pd

# os.chdir("./AI-Feynman/") #&& ./compile.sh

cart_dist = s_m.distance

polar = pd.DataFrame()
polar['sin_ra_mars'] = np.array(np.sin(np.radians(s_m.lon)))
polar['cos_ra_mars'] = np.array(np.cos(np.radians(s_m.lon)))
polar['sin_dec_mars'] = np.array(np.sin(np.radians(s_m.lat)))
polar['cos_dec_mars'] = np.array(np.cos(np.radians(s_m.lat)))
polar['mars_ang_width'] = np.array(np.arctan((0.00486759341* 0.00465047)/s_m.distance*2))
polar['sin_ra_earth'] = np.array(np.sin(np.radians(s_e.lon)))
polar['cos_ra_earth'] = np.array(np.cos(np.radians(s_e.lon)))
polar['sin_dec_earth'] = np.array(np.sin(np.radians(s_e.lat)))
polar['cos_dec_earth'] = np.array(np.cos(np.radians(s_e.lat)))
polar['earth_ang_width'] = np.array(np.arctan((0.00914913477* 0.00465047)/s_e.distance)*2)
polar['distance'] = cart_dist


polar = polar.sample(frac=1).reset_index(drop=True)

np.savetxt('./example_data/polar.txt', polar[[ "sin_ra_mars","cos_ra_mars","sin_dec_mars","cos_dec_mars","mars_ang_width",
                                                           "sin_ra_earth","cos_ra_earth","sin_dec_earth","cos_dec_earth","earth_ang_width","distance"]].sample(frac=1).values)
np.loadtxt('./example_data/polar.txt')
```
