```python
from astroquery.jplhorizons import Horizons
import numpy as np
import plotly.graph_objects as go

# Get the ephemeris for the Sun and Mar
mars = Horizons(id='499', location='675', epochs={'start': '1601-01-01', 'stop': '1602-12-31', 'step': '1d'}) #1601-12-31
sun = Horizons(id='Sun', location='675', epochs={'start': '1601-01-01', 'stop': '1602-12-31', 'step': '1d'})

mars_epth = mars.ephemerides()
sun_epth = sun.ephemerides()
```

```python
from astropy.coordinates import SkyCoord
from astropy import units as u

mars_ang_width = mars_epth['ang_width']
sun_ang_width = sun_epth['ang_width']

mars_ra = mars_epth['RA']
mars_dec = mars_epth['DEC']

sun_ra = sun_epth['RA']
sun_dec = sun_epth['DEC']
```

# Cartesian

```python
m = SkyCoord(np.array(mars_ra)*u.degree, np.array(mars_dec)*u.degree, 0.00486759341* 0.00465047 / np.tan(np.radians(mars_ang_width / 3600) / 2))
m.represent_as('cartesian')
s = SkyCoord(np.array(sun_ra)*u.degree, np.array(sun_dec)*u.degree, 0.00465047 / np.tan(np.radians(sun_ang_width / 3600) / 2))
s.represent_as('cartesian')
```

```python
import os
import pandas as pd

# os.chdir("./AI-Feynman/") #&& ./compile.sh

cart_dist = np.sqrt((s.represent_as('cartesian').x - m.represent_as('cartesian').x)**2 + (s.represent_as('cartesian').y - m.represent_as('cartesian').y)**2 + (s.represent_as('cartesian').z - m.represent_as('cartesian').z)**2)


data_cartesian = pd.DataFrame()
data_cartesian['mars_x'] = m.represent_as('cartesian').x
data_cartesian['mars_y'] = m.represent_as('cartesian').y
data_cartesian['mars_z'] = m.represent_as('cartesian').z
data_cartesian['sun_x'] = s.represent_as('cartesian').x
data_cartesian['sun_y'] = s.represent_as('cartesian').y
data_cartesian['sun_z'] = s.represent_as('cartesian').z
data_cartesian['dist'] = cart_dist

data_cartesian = data_cartesian.sample(frac=1).reset_index(drop=True)

np.savetxt('./example_data/cartesian.txt', data_cartesian[[ "mars_x","mars_y","mars_z","sun_x","sun_y","sun_z","dist"]].sample(frac=1).values)
np.loadtxt('./example_data/cartesian.txt')
```

```python
from aifeynman.S_run_aifeynman import run_aifeynman
import time

start = time.time()
run_aifeynman("./example_data/","cartesian.txt",30,"7ops.txt", bias=[1,1,1,0,1,0,1,1,1,1,1], polyfit_deg=2, NN_epochs=400)
end = time.time()
```

# Polar

```python
m = SkyCoord(np.array(mars_ra)*u.degree, np.array(mars_dec)*u.degree, 0.00486759341* 0.00465047 / np.tan(np.radians(mars_ang_width / 3600) / 2))
m.represent_as('cartesian')
s = SkyCoord(np.array(sun_ra)*u.degree, np.array(sun_dec)*u.degree, 0.00465047 / np.tan(np.radians(sun_ang_width / 3600) / 2))
s.represent_as('cartesian')
```

```python
m.distance
```

```python
import os
import pandas as pd

# os.chdir("./AI-Feynman/") #&& ./compile.sh

cart_dist = np.sqrt((s.represent_as('cartesian').x - m.represent_as('cartesian').x)**2 + (s.represent_as('cartesian').y - m.represent_as('cartesian').y)**2 + (s.represent_as('cartesian').z - m.represent_as('cartesian').z)**2)


data_cartesian = pd.DataFrame()
data_cartesian['mars_c_ra'] = np.cos(np.radians(m.ra))
data_cartesian['mars_s_ra'] = np.sin(np.radians(m.ra))
data_cartesian['mars_c_dec'] = np.cos(np.radians(m.dec))
data_cartesian['mars_s_dec'] = np.sin(np.radians(m.dec))
data_cartesian['mars_ang_width'] = np.radians(mars_ang_width / 3600) / 2
data_cartesian['sun_c_ra'] = np.cos(np.radians(s.ra))
data_cartesian['sun_s_ra'] = np.sin(np.radians(s.ra))
data_cartesian['sun_c_dec'] = np.cos(np.radians(s.dec))
data_cartesian['sun_s_dec'] = np.sin(np.radians(s.dec))
data_cartesian['sun_ang_width'] = np.radians(sun_ang_width / 3600) / 2
data_cartesian['dist'] = cart_dist


data_cartesian = data_cartesian.sample(frac=1).reset_index(drop=True)

np.savetxt('./example_data/cartesian.txt', data_cartesian[[ "mars_c_ra","mars_s_ra","mars_c_dec","mars_s_dec","mars_ang_width","sun_c_ra","sun_s_ra","sun_c_dec","sun_s_dec","sun_ang_width","dist"]].sample(frac=1).values)
np.loadtxt('./example_data/cartesian.txt')
```
