```python
def replace_equations(input_string):
  input_string = input_string.replace("exp", "np.exp")
  input_string = input_string.replace("sqrt", "np.sqrt")
  input_string = input_string.replace("pi", "np.pi")
  input_string = input_string.replace("log", "np.log")
  input_string = input_string.replace("sin", "np.sin")
  input_string = input_string.replace("anp.sin", "np.arcsin")
  input_string = input_string.replace("cos", "np.cos")
  input_string = input_string.replace("anp.cos", "np.arccos")
  input_string = input_string.replace("tan", "np.tan")
  input_string = input_string.replace("anp.tan", "np.arctan")
  return input_string
```

```python
import pandas as pd
df = pd.DataFrame(columns=["equation", "fit", "parsimony", "colour"])
```

```python
with open("/content/drive/MyDrive/mars/iiWAS2023/dimensionality/polar/Results/10var-polar2.txt", "r") as file:
  data = file.readlines()
  count=0
  for j in data:
    count+=1
    equations = j.split(' ',5)
    equation = equations[-1]
    fit = equations[0]
    parsimony = equations[-3]
    equation = replace_equations(equation.replace("\n", ""))
    if ("x0" in equation)|("x1" in equation)|("x2" in equation)|("x3" in equation)|("x4" in equation)|("x5" in equation)|("x9" in equation):
      equation = equation.replace("x0","\lambda_{M1}")
      equation = equation.replace("x1","\lambda_{M2}")
      equation = equation.replace("x2","\lambda_{M3}")
      equation = equation.replace("x3","\lambda_{M4}")
      equation = equation.replace("x4","\lambda_{M5}")
      equation = equation.replace("x5","\lambda_{S1}")
      equation = equation.replace("x6","\lambda_{S2}")
      equation = equation.replace("x7","\lambda_{S3}")
      equation = equation.replace("x8","\lambda_{S4}")
      equation = equation.replace("x9","\lambda_{S5}")
      print("Equation %s: %s, %s, %s" %(count, equation, fit, parsimony))
      entry = pd.DataFrame.from_dict({"equation": [equation], "fit": [float(fit)], "parsimony": [float(parsimony)], "colour": ["red"]})
      df = pd.concat([df, entry], ignore_index=True)
```
